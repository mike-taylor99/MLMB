```python
import os
import time
import requests
import pandas as pd
from tqdm import tqdm
```

```python
path = os.path.abspath('../data/teams.csv')
df = pd.read_csv(path)
df.head()
```

```python
SEASON = 2023
SR_SCHOOL_KEYS = [row.to_dict().get('SR key') for index, row in df.iterrows()]

get_gamelog_url = lambda school_key : f'https://www.sports-reference.com/cbb/schools/{school_key}/men/{SEASON}-gamelogs.html'
```

```python
for school in tqdm(SR_SCHOOL_KEYS):
    time.sleep(3) # Delay for 3 seconds

    url = get_gamelog_url(school)
    html = requests.get(url).content

    file_path = os.path.abspath(f'../data/seasons/{SEASON}/{school}/{school}.html')
    dir = os.path.dirname(file_path)
    if not os.path.exists(dir):
        os.makedirs(dir)

    with open(file_path, 'w') as file:
        file.write(html.decode('utf-8'))
```

```python
school = 'connecticut'
file_path = os.path.abspath(f'../data/seasons/{SEASON}/{school}/{school}.html')
team_df = pd.read_html(file_path)[0]

for index, row in team_df.iterrows():
    if index == 0:
        print(row.to_dict())

team_df.head()
```
